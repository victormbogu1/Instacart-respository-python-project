```python
# Import data
import pandas as pd
import numpy as np
import os
```

```python
# Start by creating dictionaries from the data for the months of January and February via the following code
# Define a dictionary containing January 2020 data
data1 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Jan-20', 'Jan-20', 'Jan-20', 'Jan-20'],
    'purchased_meat':[0, 13, 3, 4],
    'purchased_alcohol':[1, 2, 10, 0],
    'purchased_snacks':[10, 5, 1, 7]}
```

```python
# Define a dictionary containing February 2020 data
data2 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Feb-20', 'Feb-20', 'Feb-20', 'Feb-20'],
    'purchased_meat':[0, 10, 5, 3],
    'purchased_alcohol':[2, 4, 14, 0],
    'purchased_snacks':[15, 3, 2, 6]}
```

```python
# Convert the dictionary into dataframe
df = pd.DataFrame(data1,index=[0, 1, 2, 3])
df_1 = pd.DataFrame(data2,index=[0, 1, 2, 3])
```

```python
# you can concatenate your dataframes
frames = [df, df_1]
```

```python
df_concat = pd.concat(frames)
```

```python
# you wanted to place the two dataframes side by side (creating a wide-format dataframe), you’d need to use a different argument: axis = 1.
df_concat = pd.concat(frames, axis = 1)
```

```python
# Appending data is a straightforward approach for adding rows to an existing dataframe with the same number of columns
# All you have to be sure of is that your two dataframes have the same number of columns
df_appended = df.append(df_1)
```

```python
# Your final result will be two dataframes stacked on top of one another
df_appended
```
